```python
import re
from markitdown import MarkItDown

def eliminar_vinculos(texto):
    patron=r'\[([^\]]+)\]\([^\)]+\)'
    return re.sub(patron,r'\1',texto)

def eliminar_citas(texto):
    patron1=r'\[\[[^\]]+\]\]'
    patron2=r'\(#cite_note-[^\)]+\)'
    texto=re.sub(patron1,'',texto)
    texto=re.sub(patron2,'',texto)
    return texto

def eliminar_imagenes(texto):
    patron=r'!\[([^\]]*)\]\([^\)]+\)'
    return re.sub(patron,'',texto)

def eliminar_seccion_notas(texto):
    patron=r'## Notas[\s\S]*'
    return re.sub(patron,'',texto)

def eliminar_intro(texto):
    patron1=r'\s*◒\s*'
    texto=re.sub(patron1,'',texto,flags=re.MULTILINE)
    patron2=r'\s*# Cosmere\s*'
    texto=re.sub(patron2,'',texto,flags=re.MULTILINE)
    patron3=r'\s*De La Coppermind\s*'
    texto=re.sub(patron3,'',texto,flags=re.MULTILINE)
    patron4=r'\s*Ir a la navegación\s*'
    texto=re.sub(patron4,'',texto,flags=re.MULTILINE)
    patron5=r'\s*Ir a la búsqueda\s*'
    texto=re.sub(patron5,'',texto,flags=re.MULTILINE)
    patron6=r'\*\*La Coppermind tiene spoilers de todos los trabajos publicados de Brandon, incluyendo Viento y verdad\. La información sobre libros que aún no se han publicado \(como \*Isles of the Emberdark\*\) está permitida sólo en las páginas de los propios libros\. Para ver una visión anterior de la wiki sin spoilers de uno de estos libros, ve a nuestra Máquina del Tiempo\*\*'
    texto=re.sub(patron6,'',texto,flags=re.MULTILINE)
    patron7=r'^.*¡Esta página o sección contiene spoilers.*[\r\n]*'
    texto=re.sub(patron7,'',texto,flags=re.MULTILINE)
    patron8=r'^.*Parece que esta página necesita actualizarse.*[\r\n]*'
    texto=re.sub(patron8,'',texto,flags=re.MULTILINE)
    patron9=r'## Sumario[\s\S]*?(?=\n##|\Z)'
    texto=re.sub(patron9,'',texto,flags=re.MULTILINE)
    return texto


def html_to_md(nombre_archivo_html):

    md=MarkItDown()

    resultado=md.convert(nombre_archivo_html)

    texto=eliminar_vinculos(resultado.text_content)
    texto=eliminar_citas(texto)
    texto=eliminar_imagenes(texto)
    texto=eliminar_seccion_notas(texto)
    texto=eliminar_intro(texto)

    nombre_archivo_markdown=nombre_archivo_html.replace('.html','.md')
    with open(nombre_archivo_markdown,'w',encoding='utf-8') as f:
        f.write(texto)

#html_to_md("es.coppermind.net__wiki_Cosmere.html")   
```

```python
import os

# Obtener la lista de archivos en la carpeta actual
archivos=os.listdir()

# Filtrar solo los archivos .html
all_htmls=[archivo for archivo in archivos if archivo.endswith(".html")]

# Pasar a markdown todos los archivos .html 
for html_i in all_htmls:
    html_to_md(html_i)
```
